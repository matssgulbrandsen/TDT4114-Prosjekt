```python
import requests

# Sett base-URLen for API-et
base_url = "https://sealevel-nexus.jpl.nasa.gov"

# Endepunkt for liste over datasett
endpoint = "/list"
url = base_url + endpoint

response = requests.get(url)

if response.status_code == 200:
    data = response.json()  # API-et returnerer data i JSON-format
    print("Datasett:", data)
else:
    print("Feil ved henting av datasett:", response.status_code)

```
